**Data Cleaning:**

```python
#step 1
# Handling missing data
# For numerical columns, we can fill missing values with the mean
numerical_columns = ['BasePay', 'OvertimePay', 'OtherPay', 'Benefits']
df[numerical_columns] = df[numerical_columns].fillna(df[numerical_columns].mean())
```

```python
df.head()
```

```python
# For categorical columns
df = df.drop(['Notes', 'Status'], axis=1)
```

```python
#step 2
#remove duplicates
df_copy=df
```

```python
df_copy = df.drop_duplicates()
```

```python
df_copy.info()
```

```python
df_copy.isna().sum()
```

```python
df_copy.describe()
```

**Basic Data Visualization**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['TotalPay'])
plt.title('Boxplot of TotalPay')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df_copy['TotalPay'], bins=50)
plt.title('Distribution of TotalPay')
plt.xlabel('TotalPay')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
df_copy['JobTitle'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Proportion of Employees in Different Departments')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
df_copy['Agency'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Proportion of Employees in Different Departments')
plt.ylabel('')
plt.show()
```

```python
df_copy['Department'] = df['JobTitle'].apply(lambda title: title.split()[0].lower())


plt.figure(figsize=(8, 8))
df_copy['Department'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Proportion of Employees in Different Departments (Inferred from JobTitle)')
plt.ylabel('')
plt.show()
```

**Grouped Analysis:**

```python

grouped_data = df_copy.groupby('Department')['TotalPay'].agg(['mean', 'median', 'min', 'max', 'count'])

```

```python
grouped_data.info()
```

```python
grouped_data.head()
```

**Simple Correlation Analysis**

```python
correlation = df_copy['TotalPay'].corr(df_copy['TotalPayBenefits'])
print("Correlation between TotalPay and Benefits:", correlation)
```
